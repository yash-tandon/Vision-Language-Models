```python
import pandas as pd
import numpy as np
from PIL import Image
import os
```

```python
train_directory_path = "/kaggle/input/neu-metal-surface-defects-data/NEU Metal Surface Defects Data/train"
valid_directory_path = "/kaggle/input/neu-metal-surface-defects-data/NEU Metal Surface Defects Data/valid"
test_directory_path = "/kaggle/input/neu-metal-surface-defects-data/NEU Metal Surface Defects Data/test"
```

```python
df = []

for dirname, _, filenames in os.walk(train_directory_path):
    for filename in filenames:
        df.append([os.path.join(dirname, filename),"Metal Surface", dirname[82:]])


for dirname, _, filenames in os.walk(test_directory_path):
    for filename in filenames:
        df.append([os.path.join(dirname, filename), "Metal Surface", dirname[81:]])


for dirname, _, filenames in os.walk(valid_directory_path):
    for filename in filenames:
        df.append([os.path.join(dirname, filename), "Metal Surface", dirname[82:]])
```

```python
def clean_path_name_train(dirname,filename):
    path = os.path.join(dirname, filename)
    dirname = dirname.replace('/kaggle/input/mvtec-ad/','')
    object_type = ''
    for i in range(len(dirname)):
        if dirname[i] != '/':
            object_type = object_type + dirname[i]
        else:
            break
    defect = dirname[len(object_type)+7:]
    return [path, object_type, defect]

def clean_path_name_test(dirname,filename):
    path = os.path.join(dirname, filename)
    dirname = dirname.replace('/kaggle/input/mvtec-ad/','')
    object_type = ''
    for i in range(len(dirname)):
        if dirname[i] != '/':
            object_type = object_type + dirname[i]
        else:
            break
    defect = dirname[len(object_type)+6:]
    return [path, object_type, defect]

def clean_path_name_ground_truth(dirname,filename):
    path = os.path.join(dirname, filename)
    dirname = dirname.replace('/kaggle/input/mvtec-ad/','')
    object_type = ''
    for i in range(len(dirname)):
        if dirname[i] != '/':
            object_type = object_type + dirname[i]
        else:
            break
    defect = dirname[len(object_type)+14:]
    return [path, object_type, defect]
```

```python
path = "/kaggle/input/mvtec-ad"

for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        if "train" in os.path.join(dirname, filename):
            df.append(clean_path_name_train(dirname,filename))
        if "test" in os.path.join(dirname, filename):
            df.append(clean_path_name_test(dirname,filename))
        if "ground_truth" in os.path.join(dirname, filename):
            df.append(clean_path_name_ground_truth(dirname, filename))
```

```python
df = pd.DataFrame(df, columns = ['file_path', 'object','defect'])
df
```

```python
def generate_caption(obj, defect):
    if defect == 'good':
        return f"The {obj} does not have a defect"
    return f"The {obj} has a {defect} defect."

df["caption"] = df.apply(lambda row: generate_caption(row["object"], row["defect"]), axis=1)
```
